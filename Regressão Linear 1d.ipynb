### Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Dados

```python
X = [0,0.12,0.25,0.27,0.38,0.42,0.44,0.55,0.92,1.0]
Y = [0,0.15,0.54,0.51,0.34,0.1,0.19,0.53,1.0,0.58]
plt.scatter(X,Y)
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

### Gradient Descent

```python
#Parâmetros inicais
w1 = 1.3
w0 = 0.75
alpha = 0.01
iteracoes = 51
costs = []

for i in range(1, iteracoes):
    
#Função Custo
    Y_pred = np.multiply(w1, X) + w0 #y_chapeu
    Loss_error = 0.5 * (Y_pred - Y)**2
    cost = np.sum(Loss_error)
    
#Gradiente
    dw0 = np.sum((Y_pred - Y))
    dw1 = np.dot((Y_pred - Y), X)
    costs.append(cost)
    w1 = w1 - alpha*dw1
    w0 = w0 - alpha*dw0
    
    if i%10 == 0:
        print("Custo em", i,"iterações = ", cost)
        print("w1 =",w1, "w0 =", w0)
```

### Resultados

```python
print("w1 = ", w1,"& w0 = ",  w0)
plt.plot(costs)
plt.ylabel('Custo')
plt.xlabel('Iterações')
plt.show()
```

```python
plt.plot(X,np.multiply(w1, X) + w0)
plt.scatter(X,Y)
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```
